# Iterating over a list

```python
fruits = ["apple", "orange", "kiwi"]

for i in fruits:
    print(i)
```

```python
for char in "Charles":
    print(char)
```

```python
numbers = [1, 2, 3, 4, 5]

for n in numbers:
    if n % 2 == 0:
        print(n)
```

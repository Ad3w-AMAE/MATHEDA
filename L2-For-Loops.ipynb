# Iterating over a list

```python
fruits = ["apple", "orange", "kiwi"]

for i in fruits:
    print(i)
```

```python
numbers = [1, 2, 3, 4, 5]

for n in numbers:
    if n % 2 == 0:
        print(n)
```

```python
for j in  range(5):
    print(j)    
```

# Iterating Over a Dictionary

```python
for char in "Charles":
    print(char)
```

```python
employee = {'name':'Alice', 'age': 30, 'city': 'New York'}
```

```python
employee['city']
```

```python
for key in employee:
    print(key)
```

```python
for value in employee.values():
    print(value)
```
